```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the cycloid
r = 1  # radius of the rolling circle
theta = np.linspace(0, 4 * np.pi, 1000)  # angle in radians

# Cycloid equations
x = r * (theta - np.sin(theta))
y = r * (1 - np.cos(theta))


# Plotting the cycloid
plt.figure(figsize=(10, 5))
plt.plot(x, y, label='Cycloid', color='blue')


for i in range(0, len(x), 10):
    # Calculate the vertex of the parabola at each 10th point
    vertex_x = x[i]
    vertex_y = y[i]
    
    # Define the parabola equation: y = a * (x - h)^2 + k
    # Here, we choose a = -0.1 for a downward opening parabola
    a = -0.1
    parabola_x = np.linspace(vertex_x, vertex_x + 1, 100)  # Right side of the vertex
    parabola_y = a * (parabola_x - vertex_x) ** 2 + vertex_y
    
    # Plot the parabola
    plt.plot(parabola_x, parabola_y, color='red', alpha=0.5)



plt.title('Cycloid Generated by a Point on the Radius of a Rolling Circle')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black',linewidth=0.5, ls='--')
plt.axvline(0, color='black',linewidth=0.5, ls='--')
plt.grid()
plt.legend()
plt.axis('equal')
plt.show()
```

```python
# ... existing code ...
import ipywidgets as widgets
from IPython.display import display

# ... existing code ...

# Function to plot the cycloid with an interactive slider for 'a'
def plot_cycloid(a):
    plt.figure(figsize=(10, 5))
    plt.plot(x, y, label='Cycloid', color='blue')

    for i in range(0, len(x), 10):
        # Calculate the vertex of the parabola at each 10th point
        vertex_x = x[i]
        vertex_y = y[i]
        
        # Define the parabola equation: y = a * (x - h)^2 + k
        parabola_x = np.linspace(vertex_x, vertex_x + 1, 100)  # Right side of the vertex
        parabola_y = a * (parabola_x - vertex_x) ** 2 + vertex_y
        
        # Determine the color based on the y values
        color = 'red' if parabola_y[10] > y[i + 1] else 'blue'
        # Plot the parabola
        plt.plot(parabola_x, parabola_y, color=color, alpha=0.5) 

    plt.title('Cycloid Generated by a Point on the Radius of a Rolling Circle')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.axhline(0, color='black', linewidth=0.5, ls='--')
    plt.axvline(0, color='black', linewidth=0.5, ls='--')
    plt.grid()
    plt.legend()
    plt.axis('equal')
    plt.show()

# Create a slider for 'a'
a_slider = widgets.FloatSlider(value=-0.1, min=-1.0, max=0.0, step=0.01, description='a:')
widgets.interactive(plot_cycloid, a=a_slider)

# ... existing code ...
```
