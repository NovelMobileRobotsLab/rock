```python
# get all png files in this directory that is loss_all
import os
import glob
import numpy as np

png_files = glob.glob("loss_all*.png")
png_files.sort(key=os.path.getmtime)
print(png_files)

from PIL import Image

# Create the frames
frames = []
for i in png_files:
    new_frame = Image.open(i)
    frames.append(new_frame)

# Save into a GIF file that loops forever
frames[0].save('loss_all.gif', format='GIF',
               append_images=frames[1:],
               save_all=True,
               duration=100, loop=0)
```

```python
loss_all_hist = np.load("loss_all_hist.npy")
loss_all_hist
```

```python
import matplotlib.pyplot as plt
for i in range(len(loss_all_hist)):
    plt.hist(loss_all_hist[i], bins=100, alpha=0.1)
plt.xlabel("loss")
plt.ylabel("frequency")
plt.title("histogram of loss")
plt.show()
```
