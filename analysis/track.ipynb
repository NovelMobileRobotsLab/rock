```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib widget


df = pd.read_csv("rocktrack.csv")
df = df[df['t'] > 6]
df = df[df['t'] < 87]
#rezero df index
df = df.reset_index(drop=True)

plt.figure(figsize=(10,8))

plt.plot(df['x'], df['y'], 'orange', linewidth=3)
plt.plot(df['x'][0], df['y'][0], 'go', markersize=10)
plt.plot(df['x'][len(df)-1], df['y'][len(df)-1], 'ro', markersize=10)

plt.axis('equal')


#load image
# from PIL import Image
# img = Image.open('framestart.tiff')

# plt.g
plt.gca().xaxis.set_visible(False)
# imgplot = plt.imshow(img)
plt.axis('off')
#draw x and y axis with dotted lines
# plt.plot([0,1], [0,0], 'k--')
# plt.plot([0,0], [0,1], 'k--')

plt.savefig("track.png", transparent=True)
```
